```python
#importing libraries needed for generating data to play with
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler

```

```python
train_labels = []
train_samples = []
```

DESCRIPTION OF DATA

An experimental drug was tested on individuals from ages 13 to 65
the trial had 2100 participants. Half were under 65.
95% of patients 65 or older experienced side effects
95% if patients under 65 experienced no side effects

```python
for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
```

```python
train_samples
```

```python
#keras accepts format of a numpy array, so changing the data to the acceptable format

train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1)) #converts the data to fit into the interval 0 and 1
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1)) #using reshape, to be able to put there 1D data
```

```python
scaled_train_samples
```

Importing all the keras stuff :)

```python
import keras

from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
#constructing the model

model = Sequential([
    Dense(16, input_shape=(1,), activation='relu'),
    Dense(32, activation='relu'),
    Dense(2, activation='relu')
])
```

```python
#visualising the model for better understanding
model.summary()
```

```python
#model.compile(Optimizer)
model.compile(Adam(lr=.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#now train the model, using the .fit function

model.fit(scaled_train_samples, train_labels, validation_split = 0.1, batch_size=10, epochs=20, shuffle=True, verbose=2)
```

```python
test_labels = []
test_samples = []
```

```python
#make some test data to check how well does the model perform

for i in range(200):
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
    
for i in range(10):
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(0)
```

```python
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_test_samples = scaler.fit_transform((test_samples).reshape(-1,1))
```

```python
#Predictions

predictions = model.predict(scaled_test_samples, batch_size=10, verbose=0)
predictions

#something seems amiss
```

```python
predictions_classes = model.predict_classes(scaled_test_samples, batch_size=10, verbose=0)
predictions_classes
```

```python
#create a confusion matrix to have a representation how well our model did.
#to do so, importing the necessary packages

%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt

```
